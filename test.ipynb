```python
#Importerar moduler
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#Funktion som omvandlar CSV fil till en lista av listor som innehåller raderna i CSV filen
def read_file(file_name):
    with open(file_name, "r", encoding="utf-8-sig") as csv_file:
        dataset = []
        csv_reader = csv.reader(csv_file)
        for line in csv_reader:
            dataset.append(line)
    return dataset

#Funktion som skriver ut de 3 första raderna
def print_3_first_row(data):
    for i in range(0,3,1):
        print(data[i])

#funktion som analyserar de högsta och lägsta länderna i en kolumn
def analyze_column(dataset, column_index):
    header = dataset[0]  # Get the header row
    data = dataset[1:]  # Get the data rows

    # Sort the dataset based on the specified column
    sorted_dataset_dec = sorted(data, key=lambda x: float(x[column_index]), reverse=True)
    sorted_dataset_ace = sorted(data, key=lambda x: float(x[column_index]), reverse=False)

    # Get the top 8 and bottom 8 countries based on the specified column
    top_countries = sorted_dataset_dec[:8]
    bottom_countries = sorted_dataset_ace[:8]

    # Print the table header
    print(f"De 8 högsta värdena i kolumn {column_index} ({header[column_index]})")
    print("+-------------------------------+-----------------------------+")
    print(f"| Country name                  | {header[column_index]:<27} |")
    print("+-------------------------------+-----------------------------+")


    # Print the top countries
    for country in top_countries:
        print(f"| {country[0]:<29} | {country[column_index]:<27} |")

    # Print the table footer
    print("+-------------------------------+-----------------------------+")

    # Print the table header for the bottom countries
    print(f"De 8 lägsta värdena i kolumn {column_index} ({header[column_index]})")
    print("+-------------------------------+-----------------------------+")
    print(f"| Country name                  | {header[column_index]:<27} |")
    print("+-------------------------------+-----------------------------+")

    # Print the bottom countries
    for country in bottom_countries:
        print(f"| {country[0]:<29} | {country[column_index]:<27} |")

    # Print the table footer
    print("+-------------------------------+-----------------------------+")


    #Skapar Diagrammet
    top_n_bot_countries_names = []
    top_n_bot_countries_values = []

    for country in top_countries:
        top_n_bot_countries_names.append(country[0])
        top_n_bot_countries_values.append(float(country[column_index]))  # Convert to float

    for country in sorted(bottom_countries, key=lambda x: float(x[column_index]), reverse=True):
        top_n_bot_countries_names.append(country[0])
        top_n_bot_countries_values.append(float(country[column_index]))  # Convert to float

    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'purple', 'orange', 'pink', 'brown', 'gray', 'olive', 'cyan', 'navy', 'lime']
    plt.figure(figsize=(10, 6))
    plt.bar(top_n_bot_countries_names, top_n_bot_countries_values, color=colors)
    plt.xticks(rotation=60, fontsize=8, ha='right')

    plt.tight_layout()
    plt.show()
```

```python
def calculate_avg_WHR1Data(column_index):
    list_of_column_values = []
    for contry in WHR1Data[1:]:
        list_of_column_values.append(contry[column_index])

    sum_lista_float = 0
    for i in list_of_column_values:
        sum_lista_float =  float(i) + sum_lista_float

    objects_in_lis = len(list_of_column_values)

    avg = sum_lista_float / objects_in_lis

    avg_rounded = round(avg, 2)
    print(avg_rounded)
```
